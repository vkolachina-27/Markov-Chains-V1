```python
import math
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yfin
import random
```

```python
data = yfin.download("META", start='2018-01-01', end='2024-05-01')
data
```

```python
data["percent_change"] = 1
data = data.reset_index()
for i in np.arange(1, data.shape[0]):
  data.loc[i, "percent_change"] = data.loc[i, "Open"]/data.loc[i-1, "Open"]
```

```python
data
```

```python
plt.plot(data["Date"], data["Open"])
plt.title("Opening stock price of Meta")
plt.show()
plt.plot(data["Date"], data["percent_change"])
plt.title("Percent change of Meta")
plt.show()
```

```python
chance_up = data.loc[data["percent_change"]>=1].shape[0]/data.shape[0]
avg_up = data.loc[data["percent_change"]>=1]["percent_change"].sum()/data.loc[data["percent_change"]>=1].shape[0]

chance_down = data.loc[data["percent_change"]<1].shape[0]/data.shape[0]
avg_down = data.loc[data["percent_change"]<1]["percent_change"].sum()/data.loc[data["percent_change"]<1].shape[0]

print("Chance up", chance_up, "Chance down", chance_down)
print("Average up", avg_up, "Average down", avg_down)
```

```python
up_given_up= []
up_given_down =  []
down_given_up = []
down_given_down = []

for i in np.arange(1, data.shape[0]):

  if data.loc[i, "percent_change"] >= 1 and data.loc[i-1, "percent_change"] >=1:
    up_given_up.append(data.loc[i, "percent_change"])

  elif data.loc[i, "percent_change"] >= 1 and data.loc[i-1, "percent_change"] < 1:
    up_given_down.append(data.loc[i, "percent_change"])

  elif data.loc[i, "percent_change"] < 1 and data.loc[i-1, "percent_change"] >=1:
    down_given_up.append(data.loc[i, "percent_change"])

  else:
    down_given_down.append(data.loc[i, "percent_change"])
```

```python
chance_up_given_up = len(up_given_up)/(len(up_given_up) + len(down_given_up))
chance_up_given_down = len(up_given_down)/(len(up_given_down) + len(down_given_down))
chance_down_given_up = len(down_given_up)/(len(up_given_up) + len(down_given_up))
chance_down_given_down = len(down_given_down)/(len(down_given_down) + len(up_given_down))

print("up given up", chance_up_given_up, "down given up", chance_down_given_up, "up given down", chance_up_given_down, "down given down", chance_down_given_down)
```

```python
prob_matrix = np.array([[chance_up_given_up, chance_down_given_up], [chance_up_given_down, chance_down_given_down]])
print(prob_matrix)
```

```python
print(prob_matrix @ prob_matrix, "\n\n", prob_matrix @ prob_matrix@ prob_matrix)
```

```python
num_sims = 100
```

```python
sim_prices = []
sim_prices.append(data.loc[data.shape[0]-1, "Open"])
sim_percents = []
sim_percents.append(data.loc[data.shape[0]-1, "percent_change"])
for i in np.arange(1, num_sims):
  prob = random.uniform(0, 1)
  prev_up = (sim_percents[i-1] >= 1)
  cur_perc_change = 1;

  if prev_up and prob <= chance_up_given_up:
    cur_perc_change = avg_up
  elif prev_up:
    cur_perc_change = avg_down
  elif not prev_up and prob <= chance_down_given_down:
    cur_perc_change = avg_down
  elif not prev_up:
    cur_perc_change = avg_up

  sim_percents.append(cur_perc_change)
  sim_prices.append(cur_perc_change * sim_prices[i-1])
```

```python
plt.plot(np.arange(num_sims), sim_prices)
plt.title("Simulated price of Meta")
plt.show()
plt.plot(np.arange(num_sims), sim_percents)
plt.title("Simulated Percent change of Meta")
```

```python
def brownian_motion (start, mu, sigma, time, number):
  dt = time/number
  prices = np.zeros((number +1, 1))
  prices[0] = start

  mu_for_step = mu*dt
  sigma_for_step = np.sqrt(dt)*sigma

  for i in np.arange(1, number+1):
    prices[i] = prices[i-1] * np.exp((mu_for_step - 0.5 * sigma_for_step ** 2) * dt + sigma_for_step * math.sqrt(dt) * npr.standard_normal())
  return prices
```

```python
sample = brownian_motion(100, 0.05, 0.2, 100, 1000)
```

```python
plt.plot(np.arange(len(sample)), sample)
plt.title("Sample Geometric Brownian Motion")

plt.ylabel("Simulated Asset Price")
plt.xlabel("Time Step")
```

```python
def monte_carlo (start, mu, sigma, time, number, number_iterations):
  ending_price = np.zeros((number_iterations, 1))
  for i in np.arange(number_iterations):
    ending_price[i] = brownian_motion(start, mu, sigma, time, number)[-1]
  return ending_price
```

```python
ending_p_sim = monte_carlo (100, 0.05, 0.2, 100, 1000, 1000)
```

```python
plt.hist(ending_p_sim, bins=50)
plt.title("histogram of simulated asset")
plt.ylabel("count")
plt.xlabel("simulated final price")
```
